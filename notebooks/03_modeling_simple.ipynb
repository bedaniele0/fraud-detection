```python
"""
Modelado Simple - DetecciÃ³n de Fraude
Autor: Ing. Daniel Varela Perez
"""
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix
import joblib
import warnings
warnings.filterwarnings('ignore')

print("âœ… Imports completados")
```

```python
# Cargar datos procesados
print("ğŸ“‚ Cargando datos...")

train_df = pd.read_parquet('../data/processed/train_clean.parquet')
test_df = pd.read_parquet('../data/processed/test_clean.parquet')

print(f"Train: {train_df.shape}")
print(f"Test: {test_df.shape}")
print("âœ… Datos cargados")
```

```python
# Separar features y target
feature_cols = [col for col in train_df.columns if col not in ['Class', 'Time']]
target_col = 'Class'

X_train = train_df[feature_cols]
y_train = train_df[target_col]

X_test = test_df[feature_cols]
y_test = test_df[target_col]

print(f"Features: {len(feature_cols)}")
print(f"Fraudes train: {y_train.sum()} ({y_train.sum()/len(y_train)*100:.4f}%)")
print(f"Fraudes test: {y_test.sum()} ({y_test.sum()/len(y_test)*100:.4f}%)")
print("âœ… Datos preparados")
```

```python
# Entrenar modelo Random Forest
print("ğŸš€ Entrenando Random Forest...")

rf_model = RandomForestClassifier(
    n_estimators=100,
    random_state=42,
    class_weight='balanced',
    n_jobs=-1
)

rf_model.fit(X_train, y_train)
print("âœ… Modelo entrenado")
```

```python
# ğŸ§  Entrenamiento del modelo Random Forest

from sklearn.ensemble import RandomForestClassifier

# Crear el modelo
rf_model = RandomForestClassifier(
    n_estimators=100,
    max_depth=10,
    random_state=42,
    class_weight='balanced',
    n_jobs=-1
)

# Entrenar el modelo
print("ğŸ§ª Entrenando modelo Random Forest...")
rf_model.fit(X_train, y_train)

# Calcular F1-Score en test
from sklearn.metrics import f1_score
y_pred_temp = rf_model.predict(X_test)
f1 = f1_score(y_test, y_pred_temp)

print(f"âœ… Entrenamiento completado - F1-Score: {f1:.4f}")
```

```python
# ğŸ§  Entrenamiento del modelo Random Forest

from sklearn.ensemble import RandomForestClassifier

# Crear el modelo
rf_model = RandomForestClassifier(
    n_estimators=100,
    max_depth=10,
    random_state=42,
    class_weight='balanced',
    n_jobs=-1
)

# Entrenar el modelo
print("ğŸ§ª Entrenando modelo Random Forest...")
rf_model.fit(X_train, y_train)

# Calcular F1-Score en test
from sklearn.metrics import f1_score
y_pred_temp = rf_model.predict(X_test)
f1 = f1_score(y_test, y_pred_temp)

print(f"âœ… Entrenamiento completado - F1-Score: {f1:.4f}")
```

```python
# Hacer predicciones
print("ğŸ¯ Haciendo predicciones...")

y_pred = rf_model.predict(X_test)
y_proba = rf_model.predict_proba(X_test)[:, 1]

print(f"Predicciones: {len(y_pred)}")
print(f"Fraudes predichos: {y_pred.sum()}")
print("âœ… Predicciones completadas")
```

```python
# Evaluar modelo
print("ğŸ“Š EVALUACIÃ“N DEL MODELO")
print("=" * 30)

# Classification Report
print("\nClassification Report:")
print(classification_report(y_test, y_pred))

# Confusion Matrix
print("\nConfusion Matrix:")
cm = confusion_matrix(y_test, y_pred)
print(cm)

# MÃ©tricas principales
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

print(f"\nğŸ“ˆ MÃ‰TRICAS PRINCIPALES:")
print(f"Accuracy: {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1-Score: {f1:.4f}")

print("\nâœ… EvaluaciÃ³n completada")
```

```python
# Guardar modelo
model_path = '../models/simple_fraud_model.pkl'
joblib.dump(rf_model, model_path)

print(f"ğŸ’¾ Modelo guardado: {model_path}")
print(f"ğŸ† Mejor F1-Score: {f1:.4f}")
print("âœ… Proceso completado")

# Resumen final
print(f"\nğŸ¯ RESUMEN FINAL:")
print(f"â€¢ Algoritmo: Random Forest")
print(f"â€¢ Features: {len(feature_cols)}")
print(f"â€¢ Train size: {len(X_train):,}")
print(f"â€¢ Test size: {len(X_test):,}")
print(f"â€¢ F1-Score: {f1:.4f}")
print(f"â€¢ Modelo guardado exitosamente")

print(f"\nğŸ‘¨â€ğŸ’» Desarrollado por: Ing. Daniel Varela Perez")
print(f"ğŸ“§ bedaniele0@gmail.com | ğŸ“± +52 55 4189 3428")
```
